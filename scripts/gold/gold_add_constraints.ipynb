# Add Contraints

```python
%sql
--Drop Old Constraints
ALTER TABLE workspace.gold.dim_customers
DROP CONSTRAINT pk_dim_customers CASCADE;

ALTER TABLE workspace.gold.dim_products
DROP CONSTRAINT pk_dim_products CASCADE;

ALTER TABLE workspace.gold.fact_sales
DROP CONSTRAINT IF EXISTS fk_sales_customer;
--Make Columns Non Nullable
ALTER TABLE workspace.gold.dim_customers
ALTER COLUMN customer_key SET NOT NULL;

ALTER TABLE workspace.gold.dim_products
ALTER COLUMN product_key SET NOT NULL;

-- Primary Keys
ALTER TABLE workspace.gold.dim_customers
ADD CONSTRAINT pk_dim_customers PRIMARY KEY (customer_key);

ALTER TABLE workspace.gold.dim_products
ADD CONSTRAINT pk_dim_products PRIMARY KEY (product_key);

-- Foreign Keys
ALTER TABLE workspace.gold.fact_sales
ADD CONSTRAINT fk_sales_customer
FOREIGN KEY (customer_key) REFERENCES workspace.gold.dim_customers(customer_key);

ALTER TABLE workspace.gold.fact_sales
ADD CONSTRAINT fk_sales_product
FOREIGN KEY (product_key) REFERENCES workspace.gold.dim_products(product_key);
```
